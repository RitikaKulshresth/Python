### Memory Management - Continued.....

When a Python program is executed, there are different types of areas which get created in the RAM.




- Code Area: Code gets loaded in this area. It is fixed in size.
- Variable area: In this area, variables are stored and they refer their corresponding values in heap area.
- Heap Area: Stores valuesin random order. This area may grow or shrink dynamically. Every value which gets stored here will have a specific ID/memory. Garbage collector runs in heap area from time to time and destroys the object which are not having any reference.
- Stack Area: It may grow or shrink dynamically and is responsible for the order in which method will be resolved.


### Memory de-allocation

- Garbage collector(gc) is used for memory deallocation.
- gc resides in the heap area.
- To understand the process of garbage collection, we should understand reference counting.
- Reference couting:
    - it counts the number of times a value has been referenced.
    - in other words, how many variables are refering to a particular memory address where a value is stored.
    
    


```python
a = 1000
```

```python
# id(var) => function to fetch the memory address where the variable 'var' is referring
```

```python
id(a)
```

- At this point, the reference count of the memory address '2547617711408' becomes 0.
- The value (1000) is NOT required as there is no variable referring to it.
- Garbage collector sees that the reference count is 0 and it deletes the value from that memory address and reclaims it.
- This process is called Garbage collection.

- GC is a process of deleting a value from the memory address whose reference count becomes 0 to free up that memory for storing something else.
- GC in Python is an automatic process.
