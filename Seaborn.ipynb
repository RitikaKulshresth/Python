```python
# Outliers
# Values which are beyond the normal range
# Values which are far away from the mean value
```

Boxplot uses the below formula to find outliers

1. Find Q1
2. Find Q2
3. Find Interquartile range (IQR) = Q3 - Q1
4. Find upper and lower limits/bounds
    - lower bound = Q1 - IQR * 1.5
    - upper bound = Q3 + IQR * 1.5

```python
q1 = tips_df['tip'].quantile(.25)
```

```python
q1
```

```python
q3 = tips_df['tip'].quantile(.75)
```

```python
q3
```

```python
iqr = q3 - q1
```

```python
iqr
```

```python
lower_bound = q1 - iqr * 1.5
```

```python
lower_bound
```

```python
upper_bound = q3 + 1.5 * iqr
```

```python
upper_bound
```

```python
# Any value which is more than upper_bound and less than lower_bound will be treated as outlier by boxplot
```
