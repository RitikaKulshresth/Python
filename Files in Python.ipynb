```python
# Write a python program to find below output using loop:-
# Input:- 'This is Python class', sep = ' is',
# Output:- ['This', 'is', 'Python class']


result = []
start = 0
str1 = str(input("Enter String: "))
sep1 = str(input("Enter sepreter String: "))

while True :
    index = str1.find(sep1,start)
    if index == -1 :
        
        result.append(str1[start:])
        break

    result.append(str1[start:index].strip())
    result.append(sep1.strip())
    start = index + len(sep1)
print(result)
```
